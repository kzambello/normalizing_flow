```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt

import os
import re

import sklearn
import sklearn.datasets
import sklearn.model_selection
from sklearn.datasets import fetch_openml
from sklearn.preprocessing import StandardScaler, MinMaxScaler

import tensorflow as tf

# Use keras 2
import tf_keras as tfk
os.environ["TF_USE_LEGACY_KERAS"] ="1"

## Use keras 3
#tfk = tf.keras

from normalizing_flow import MAF
```

```python
os.environ["CUDA_VISIBLE_DEVICES"]="0"
```

```python
def logit(x):
    return np.log(x / (1 - x))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def ohe(x):
    vec = np.zeros(10)
    vec[x] = 1.0
    return vec
```

```python
nvars = 784
ncondvars = 10
nflow = MAF(nvars, ncondvars, nblocks=5, hidden_units=[1024, 1024, 1024], activation='relu', last_activation='relu', l1=0.0, l2=1.0e-6)

nflow.load_weights('my_model.weights')

#print(nflow.trainable_variables)

#nflow.compile(optimizer=tfk.optimizers.Adam(learning_rate=1.0e-4, clipnorm=0.01), loss=lambda _, log_prob: -log_prob)
```

```python
nsamples = 160

cond = 0
cond = ohe(cond)

samples = nflow.sample(cond, nsamples)

samples = sigmoid(samples)*256.0

plt.figure()
for n in np.arange(16):
    plt.subplot(4,4,1+n)
    sample = samples[n].reshape((28,28))
    plt.imshow(sample, interpolation='nearest')
    #plt.imshow(sample)
    
plt.savefig("images/mnist-"+str(np.nonzero(cond)[0][0])+".png")
```
